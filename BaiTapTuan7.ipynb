```python
def binary_search(arr, value):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else:
            right = mid - 1 

    return -1

array = list(map(int, input().split()))
value = int(input())

binary_search(array, value)
```

```python
def count_occurrences(arr, x) -> int:
    count = 0
    for i in arr:
        if i == x:
            count +=1
    return count

array = list(map(int, input().split()))
value = int(input())

count_occurrences(array, value)
```
