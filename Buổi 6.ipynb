```python
input_string = input()
pairs = input_string.split()
result_dict = {}
for pair in pairs:
    if ':' in pair:
        key, value = pair.split(':', 1)
        if key not in result_dict:
            result_dict[key] = [value]
        else:
            result_dict[key].append(value)

print(result_dict)
```

```python
n1 = input()
n2 = input()
def creat_dct(n):
    lst = n.split()
    result_dict = {}
    for i in lst:
        if ':' in i:
            key, value = i.split(':', 1)
        if key not in result_dict:
            result_dict[key] = int(value)
    return result_dict

n12 = creat_dct(n1)
n22 = creat_dct(n2)

for key, value in n22.items():
    if key in n12:
        n12[key] += value
    else:
        n12[key] = value

print(n12)

```

```python
arr = list(map(int, input().split()))
k = int(input())

find= set()

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == k:
            find.add((arr[i], arr[j]))

result_list = sorted(list(find), key=lambda x: x[0])

print(result_list)
```
