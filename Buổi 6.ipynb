```python
n = list(map(int, input().split()))
n.sort()
a = []
for i in range(len(n) - 1):
    if n[i] != n[i + 1]:
        a.append(n[i])
if n[len(n) - 2] != n[len(n) - 1]:
    a.append(n[len(n) - 1])
print(a)
```

```python
n = list(map(int, input().split()))
a = []
for i in range(len(n)):
    if i == 0:
        a.append(n[i])
    else:
        a.append(a[i - 1] + n[i])
print(a)
```

```python
def left_rotate_tuple():
    line1 = input()
    tup = tuple(map(int, line1.split()))
    k = int(input())
    rotated_tuple = tup[k:] + tup[:k]
    print(rotated_tuple)

left_rotate_tuple()
```

```python
input_string = input()
pairs = input_string.split()
result_dict = {}
for pair in pairs:
    if ':' in pair:
        key, value = pair.split(':', 1)
        if key not in result_dict:
            result_dict[key] = [value]
        else:
            result_dict[key].append(value)

print(result_dict)
```

```python
list_int = list(map(int, input().split()))
pos = 0
neg = 0
zero = 0
for i in list_int:
    if i > 0:
        pos += 1
    elif i < 0:
        neg += 1
    else:
        zero += 1
dct_int = {"positive": pos, "negative": neg, "zero": zero}
print(dct_int)
```

```python
lst = list(map(str, input().split()))
dct = {}
for key in lst:
    if key not in dct:
        dct[key] = 1
    else:
        dct[key] += 1
print(dct)
```
