```python
n = list(map(int, input().split()))
n.sort()
a = []
for i in range(len(n) - 1):
    if n[i] != n[i + 1]:
        a.append(n[i])
if n[len(n) - 2] != n[len(n) - 1]:
    a.append(n[len(n) - 1])
print(a)
```

```python
n = list(map(int, input().split()))
a = []
for i in range(len(n)):
    if i == 0:
        a.append(n[i])
    else:
        a.append(a[i - 1] + n[i])
print(a)
```

```python
def left_rotate_tuple():
    line1 = input()
    tup = tuple(map(int, line1.split()))
    k = int(input())
    rotated_tuple = tup[k:] + tup[:k]
    print(rotated_tuple)

left_rotate_tuple()
```

```python
input_string = input()
pairs = input_string.split()
result_dict = {}
for pair in pairs:
    if ':' in pair:
        key, value = pair.split(':', 1)
        if key not in result_dict:
            result_dict[key] = [value]
        else:
            result_dict[key].append(value)

print(result_dict)
```

```python
list_int = list(map(int, input().split()))
dct_int = {"positive": 0, "negative": 0, "zero": 0}
for i in list_int:
    if i > 0:
        dct_int["positive"] += 1
    elif i < 0:
        dct_int["negative"] += 1
    else:
        dct_int["zero"] += 1

print(dct_int)
```

```python
n1 = input()
n2 = input()
def creat_dct(n):
    lst = n.split()
    result_dict = {}
    for i in lst:
        if ':' in i:
            key, value = i.split(':', 1)
        if key not in result_dict:
            result_dict[key] = int(value)
    return result_dict

n12 = creat_dct(n1)
n22 = creat_dct(n2)

for key, value in n22.items():
    if key in n12:
        n12[key] += value
    else:
        n12[key] = value

print(n12)

```

```python
arr = list(map(int, input().split()))
k = int(input())

find= set()

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == k:
            find.add((arr[i], arr[j]))

result_list = sorted(list(find), key=lambda x: x[0])

print(result_list)
```
