```python
def binary_search(arr, target):
    left , right = 0, len(arr)-1
    while left <= right :
        mid = (left + right) //2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            right = mid -1
        else:
            left = mid + 1
arr=list(map(int, input().split()))
target=int(input())
print(binary_search(arr,target)) 
```

```python
#Bubble sort:
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
#Selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(0, n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
li = list(map(int, input().split()))
selection_sort(li)
print(li)
    
    
```

```python
def so_lan(lst):
    d={}
    for i in lst:
        if i not in d:
            d[i] =1
        else:
            d[i] +=1
    max_value=0
    min_indx=0
    for key, value in d.items():
        if value > max_value:
            max_value=value
            min_indx = key
    print(f"{min_indx} xuat hien nhieu nhat, som nhat, {max_value} lan")
lst=list(map(int, input().split()))
so_lan(lst)
```

```python
def so_cap(lst,x):
    n=len(lst)
    count=0
    for i in range(n):
        for i in range(i+1,n):
            if lst[i] + lst[j] == x:
                count +=1
    return count
lst=list(map(int, input().split()))
arr=int(input())
print(so_cap(lst,arr))
```

```python
def day_tang(lst):
    n = len(lst)
    if n == 0:
        return 0
    dp = [1] * n 
    for i in range(n):
        for j in range(i):
            if lst[j] < lst[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
lst = eval(input())
print(day_tang(lst))
```

```python
def bin_se(x, pa):
    le = 0
    ri = len(pa) - 1
    
    ans = -1
    while le <= ri:
        mid = (le + ri) // 2
        if x <= pa[mid][1]:
            if x >= pa[mid][0]:
                ans = pa[mid][1] - pa[mid][0] + 1
                ri = mid - 1
            else: le = mid + 1
        else:
            le = mid + 1
    return ans


pa = eval(input())
pa.sort(key = lambda x: (x[1], x[1] - x[0] + 1))
li = eval(input())
for x in li:
    print(bin_se(x, pa), end = " ")
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def xuli(node, val, m):
    ans = 0
    while True:
        if val < node.val:
            ans = max(ans, m - (node.val - val))
            if node.left == None:
                node.left = Node(val)
                break
            else:
                node = node.left
        else:
            ans = max(ans, val)
            if node.right == None:
                node.right = Node(val)
                break
            else: 
                node = node.right
        
    return ans

li = list(map(int, input().split()))
m = int(input())
s = li[0] % m
root = Node(s)
ans = 0
for i in range(len(li)):
    s += li[i]
    s %= m
    ans = max(ans, xuli(root, s, m))
print(ans)
```
