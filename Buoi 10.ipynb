```python
#indexerror
my_list = list(map(int, input().split()))
n = int(input())
try:
    print(my_list[n])
except IndexError:
    print("Truy cập chỉ số sai")
```

```python
#valueError
def happy(n):
    infinity = set()
    while n != 1 and n not in infinity:
        infinity.add(n)
        n = sum(int(i) ** 2 for i in str(n))
    if n == 1:
        print('yes')
    else:
        print('no')
n = int(input())
happy(n)
```

```python
lst = list(input().split())
newlst = []
for i in lst:
    try:
        newlst.append(int(i))
    except ValueError:
        print(f"{i} không thể chuyển được")
print(newlst)
```
